```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, average_precision_score

# Define IoU (Intersection over Union) function
def calculate_iou(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])
    intersection = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)
    iou = intersection / float(box1_area + box2_area - intersection)
    return iou

# Define function to calculate average precision
def calculate_average_precision(tp, fp, num_ground_truth):
    tp_fp = np.concatenate([tp, fp], axis=0)
    sorted_indices = np.argsort(-tp_fp[:, -1])
    tp_fp_sorted = tp_fp[sorted_indices]

    cumsum_tp = np.cumsum(tp_fp_sorted[:, 0])
    cumsum_fp = np.cumsum(1 - tp_fp_sorted[:, 0])

    precision = cumsum_tp / (cumsum_tp + cumsum_fp)
    recall = cumsum_tp / num_ground_truth

    # Calculate interpolated precision at each recall point
    max_precision = np.maximum.accumulate(precision[::-1])[::-1]  # Compute max precision for decreasing recall
    average_precision = np.sum((recall[1:] - recall[:-1]) * max_precision[:-1])

    return average_precision


# Initialize lists to store ground truth and predicted boxes
gt_boxes_list = []
pred_boxes_list = []

# Iterate through the test dataset to get ground truth boxes and predictions
for i in range(len(Test_dataset)):
    image, target = Test_dataset[i]
    gt_boxes = target['boxes'].numpy()
    gt_boxes_list.extend(gt_boxes)

    # Predict boxes for each image
    predictions = model.predict(image)
    labels, boxes, scores = predictions
    pred_boxes_list.extend(boxes)

# Convert lists to numpy arrays
gt_boxes_array = np.array(gt_boxes_list)
pred_boxes_array = np.array(pred_boxes_list)

# Initialize lists to store true positives, false positives, and false negatives
tp = []
fp = []

# Loop through each predicted box
for pred_box in pred_boxes_array:
    iou_list = [calculate_iou(pred_box, gt_box) for gt_box in gt_boxes_array]
    if iou_list:  # Check if iou_list is not empty
        max_iou = max(iou_list)
        if max_iou >= 0.5:
            tp.append(np.append([1], pred_box))  # Append 1 for true positive
            gt_boxes_array = np.delete(gt_boxes_array, np.argmax(iou_list), axis=0)
        else:
            fp.append(np.append([0], pred_box))  # Append 0 for false positive
    else:
        fp.append(np.append([0], pred_box))  # Append 0 for false positive

# Remaining ground truth boxes are false negatives
fn = np.hstack((np.ones((len(gt_boxes_array), 1)), gt_boxes_array))  # Append 1 for false negative

# Calculate precision and recall
precision = len(tp) / (len(tp) + len(fp))
recall = len(tp) / (len(tp) + len(fn))

# Calculate average precision (AP)
average_precision = calculate_average_precision(np.array(tp), np.array(fp), len(gt_boxes_list))

print("Precision:", precision)
print("Recall:", recall)
print("Average Precision:", average_precision)
```
